```python
from bs4 import BeautifulSoup
import requests


res = requests.get('https://search.shopping.naver.com/best100v2/main.nhn')
soup = BeautifulSoup(res.content, 'html.parser')
#CSS Selector로 검색
items = soup.select('#popular_srch_lst > li > span.txt') #copy selector를 활용
for item in items :    
    print(item.get_text())    
```

```python
#주소를 list data에 저장하여 일시 검색
site_list = ['https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000000','https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000003']

for site in site_list:
    res = requests.get(site)
    soup = BeautifulSoup(res.content, 'html.parser')
    items = soup.select('#productListArea > ul > li > p > a')
    print(site)
    for item in items:
        print(item.get_text())
```
