```python
import requests
import sys
import openpyxl

sys.path.append('../')
from NaverAPI import naver_key
client_key, client_secret = naver_key()
start, num= 1,0
search_item = '그래픽카드'

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.column_dimensions['B'].width = 100
excel_sheet.column_dimensions['C'].width = 100
excel_sheet.append(['랭킹','제목','최저가','링크'])

for index in range(10):
    start_number = start +(index*100)
    naver_open_api = f'https://openapi.naver.com/v1/search/shop.json?query={search_item}&display=100&start=' + str(start_number)
    header_params = {"X-Naver-Client-Id":client_key,"X-Naver-Client-Secret":client_secret}
    res = requests.get(naver_open_api, headers=header_params)
    if res.status_code ==200:
        data = res.json()
        for item in data['items']:
            num+=1
            excel_sheet.append([num, item['title'],item['lprice'], item['link']])
    else :
        print ("Error!!", res_status_code)    
excel_file.save(f'shopping_with_{search_item}.xlsx')
excel_file.close()
```
